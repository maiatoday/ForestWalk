```kotlin
import net.maiatoday.forestwalk.lsystem.plant
import net.maiatoday.forestwalk.lsystem.Plant

val p1 = plant {
    seed = "F"
    rules = mapOf(
        "F" to "FF-[-F+F+F]+[+F-F-F]"
    )
    grow(3)
}

val p2 = plant {
    seed = "FF"
    rules = mapOf(
        "F" to "FF-[-F+F+F]+[+F-F-F]",
        // "FF" to "F"
    )
    grow(4)
}

val plantList = mutableListOf<Plant>()
plantList.add(p1)
plantList.add(p1)
plantList.add(p1)
plantList.add(p2)
plantList.add(p2)
plantList.add(p2)
plantList.add(p2)

val dff = plantList.toDataFrame()
dff.writeCSV("nursery.csv")

dff
```
