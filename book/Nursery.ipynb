```kotlin
%use dataframe
import net.maiatoday.forestwalk.lsystem.plant
import net.maiatoday.forestwalk.lsystem.Plant
import net.maiatoday.forestwalk.lsystem.PlantBuilder

val treeBuilder: PlantBuilder.() -> Unit = {
    seed = "X"
    rules = mapOf(
        "X" to "F[+X][-X]FX",
        "F" to "FF"
    )
    sprout(5, listOf(1.0, 0.9))
}

val shrubBuilder: PlantBuilder.() -> Unit = {
    seed = "F"
    rules = mapOf(
        "F" to "FF-[-F+F+F]+[+F-F-F]"
    )
    grow(4)
}

val plantList = buildList<Plant> { 
    repeat(25) {
        this.add(plant(treeBuilder))
    }
    repeat(20) {
        this.add(plant(shrubBuilder))
    }
}

val dff = plantList.toDataFrame()
dff.writeCSV("nursery.csv")
dff
```

```kotlin
import kotlin.random.Random

enum class Ground { Fore, Middle, Back }

val blueprints = dff
    .add("startX") {
        Random.nextInt(0, 1000)
    }
    .add("startY") {
        Random.nextInt(650, 700)
    }
    .add("isTree") {
        rules.size > 1
    }
    .add("ground") { Ground.Fore }
    .remove { seed and rules and iterations }
blueprints
```

```kotlin
import Line_8_jupyter.Ground

data class DisplayPlant(
    val bluePrint: String,
    val startX: Int,
    val startY: Int,
    val isTree: Boolean,
    val ground: Ground
) {
    val colors = if (isTree) treeShades else shrubShades
    val length = if (isTree) 5 else 6
    val startStroke = if (isTree) 15.0f else 8.0f
    val strokeChange = if (isTree) 0.98f else 0.9f
}

val forest = blueprints.toListOf<DisplayPlant>()

val forestImage = canvas(1000, 700) { width, height ->
    skyEarth(width, height, springSky, springEarth)
    forest.shuffled().forEach {
        goSquirrel(
            startX = it.startX,
            startY = it.startY,
            angleChangeDeg = 25.0,
            startStroke = it.startStroke,
            strokeChange = it.strokeChange,
            length = it.length,
            bluePrint = it.bluePrint,
            penColors = it.colors
        )
    }
}
canvas(1000, 700, startImage = forestImage) { width, height ->
    repeat(2000) {
        val p = randomOffset(width = width, height = height, startY = 300)
        val radius = randomRadius(2, 6)
        color = randomFrom(petalShades, ((p.y / height.toFloat()) * 255).toInt())
        drawOval(p.x, p.y, radius, radius)
    }
}


```
